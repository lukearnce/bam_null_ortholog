```python
# Standard library packages
import os

# Import Numpy, Pandas and Seaborn
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#Load blast results into dataframe

blast_results = pd.read_csv('/Users/la424/Aquadro_lab/fasta_files_blast_database/blast_outputs/blast_lcup.outfmt6', sep="\t", header=None)
blast_reciprocal_results = pd.read_csv('/Users/la424/Aquadro_lab/fasta_files_blast_database/blast_outputs/blast_reciprocal_lcup.outfmt6', sep="\t", header=None)

headers = ["query", "subject", "identity", "length", "mismatch", "gapopen", "qstart", "qend", "sstart", "send", "evalue", "bitscore"]
blast_results.columns = headers
blast_reciprocal_results.columns = headers
```

```python
#Merge forward and reciprocal results

rbbh = pd.merge(blast_results, blast_reciprocal_results[['query', 'subject']], left_on='subject', right_on='query', how='outer')
```

```python
#Discard rows that are not RBH
rbbh = rbbh.loc[rbbh.query_x == rbbh.subject_y]
```

```python
# Group duplicate RBH rows, taking the maximum value in each column
rbbh = rbbh.groupby(['query_x', 'subject_x']).max()
```
